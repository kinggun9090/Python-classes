````python
# @title ì™„ì„±ì½”ë“œ

import random

def generate_lotto():
    """ë¡œë˜ ë²ˆí˜¸ 6ê°œë¥¼ ìƒì„±í•˜ëŠ” í•¨ìˆ˜"""
    # random.sampleë¡œ ì¤‘ë³µ ì—†ì´ 6ê°œ ì„ íƒ
    numbers = random.sample(range(1, 46), 6)  # 1~45ì—ì„œ 6ê°œ ì„ íƒ
    numbers.sort()  # ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
    return numbers

print("ğŸ° ë¡œë˜ ë²ˆí˜¸ ìƒì„±ê¸°")
print("=" * 30)

# ì‚¬ìš©ìê°€ ì›í•˜ëŠ” ë§Œí¼ ìƒì„±
while True:
    try:
        count = int(input("ëª‡ ê°œì˜ ë²ˆí˜¸ë¥¼ ìƒì„±í• ê¹Œìš”? (0ì€ ì¢…ë£Œ): "))

        if count == 0:
            print("í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤. ğŸ‘‹")
            break
        elif count < 0:
            print("ì–‘ìˆ˜ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”!")
            continue

        print(f"\nğŸ¯ {count}ê°œì˜ ë¡œë˜ ë²ˆí˜¸:")
        print("-" * 25)

        for i in range(count):
            lotto_numbers = generate_lotto()
            print(f"{i+1}ë²ˆì§¸: {lotto_numbers}")
            # ì˜ˆìƒ ê²°ê³¼: [3, 15, 22, 31, 38, 44] (ë§¤ë²ˆ ë‹¤ë¦„)

        print(f"\nì´ {count}ê°œì˜ ë²ˆí˜¸ê°€ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤!")
        print("í–‰ìš´ì„ ë¹•ë‹ˆë‹¤! ğŸ€\n")
        print("í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤. ğŸ‘‹")
        break

    except ValueError:
        print("ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”!")
````

## ğŸ”„ ìƒì† (Inheritance)

**ìƒì†ì€** ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ì†ì„±ê³¼ ë©”ì†Œë“œë¥¼ ìì‹ í´ë˜ìŠ¤ê°€ ë¬¼ë ¤ë°›ëŠ” ê±¸ ì˜ë¯¸í•©ë‹ˆë‹¤.

- **`super()`**: ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ë©”ì†Œë“œë‚˜ ì†ì„±ì— ì ‘ê·¼í•  ë•Œ ì‚¬ìš©
- **ì˜¤ë²„ë¼ì´ë”©(Overriding)**: ìì‹ í´ë˜ìŠ¤ì—ì„œ ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ë©”ì†Œë“œë¥¼ ì¬ì •ì˜í•˜ëŠ” ê²ƒ

````python
# ë¶€ëª¨ í´ë˜ìŠ¤
class Employee:
    def __init__(self, name, salary):
        self.name = name        # ì´ë¦„
        self.salary = salary    # ê¸‰ì—¬

    def introduce(self):
        print(f"ì•ˆë…•í•˜ì„¸ìš”, {self.name}ì…ë‹ˆë‹¤.")

    def work(self):
        print(f"{self.name}ì´(ê°€) ì—…ë¬´ë¥¼ í•©ë‹ˆë‹¤.")

# ìì‹ í´ë˜ìŠ¤ 1
class Developer(Employee):
    def __init__(self, name, salary, language):
        super().__init__(name, salary)  # ë¶€ëª¨ì˜ __init__ í˜¸ì¶œ
        self.language = language        # ì‚¬ìš© ì–¸ì–´

    def work(self):  # ë©”ì†Œë“œ ì˜¤ë²„ë¼ì´ë”©
        print(f"{self.name}ì´(ê°€) {self.language}ë¡œ ì½”ë”©í•©ë‹ˆë‹¤. ğŸ’»")

# ìì‹ í´ë˜ìŠ¤ 2
class Designer(Employee):
    def __init__(self, name, salary, tool):
        super().__init__(name, salary)  # ë¶€ëª¨ì˜ __init__ í˜¸ì¶œ
        self.tool = tool                # ì‚¬ìš© ë„êµ¬

    def work(self):  # ë©”ì†Œë“œ ì˜¤ë²„ë¼ì´ë”©
        print(f"{self.name}ì´(ê°€) {self.tool}ë¡œ ë””ìì¸í•©ë‹ˆë‹¤. ğŸ¨")
````

### ğŸ”§ ë©”ì†Œë“œ íƒ€ì… ë¹„êµ

| êµ¬ë¶„ | ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ | í´ë˜ìŠ¤ ë©”ì†Œë“œ | ìŠ¤íƒœí‹± ë©”ì†Œë“œ |
| --- | --- | --- | --- |
| **ë°ì½”ë ˆì´í„°** | ì—†ìŒ | `@classmethod` | `@staticmethod` |
| **ì²« ë²ˆì§¸ ë§¤ê°œë³€ìˆ˜** | `self` | `cls` | ì—†ìŒ |
| **ì ‘ê·¼ ê°€ëŠ¥í•œ ë³€ìˆ˜** | ì¸ìŠ¤í„´ìŠ¤ + í´ë˜ìŠ¤ ë³€ìˆ˜ | í´ë˜ìŠ¤ ë³€ìˆ˜ë§Œ | ì—†ìŒ |
| **í˜¸ì¶œ ë°©ë²•** | `ê°ì²´.ë©”ì†Œë“œ()` | `í´ë˜ìŠ¤.ë©”ì†Œë“œ()` | `í´ë˜ìŠ¤.ë©”ì†Œë“œ()` |

````python
class Student:
    # í´ë˜ìŠ¤ ë³€ìˆ˜
    school_name = "ì²­ë‹´êµ­ì œê³ ë“±í•™êµ"     # í•™êµëª…
    total_students = 0           # ì „ì²´ í•™ìƒ ìˆ˜

    def __init__(self, name, grade):
        # ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜
        self.name = name         # í•™ìƒ ì´ë¦„
        self.grade = grade       # í•™ë…„
        self.scores = []         # ê°œì¸ ì„±ì 
        Student.total_students += 1

    # ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ (ê°œì¸ ì„±ì  ê´€ë¦¬)
    def add_score(self, subject, score):
        self.scores.append(f"{subject}: {score}ì ")
        return f"{self.name}ì˜ {subject} ì„±ì ì´ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤."

    def get_average(self):
        if not self.scores:
            return 0
        total = sum(int(score.split(': ')[1].replace('ì ', '')) for score in self.scores)
        return total / len(self.scores)

    # í´ë˜ìŠ¤ ë©”ì†Œë“œ (í•™êµ ì „ì²´ ê´€ë¦¬)
    @classmethod
    def get_student_count(cls):
        return f"í˜„ì¬ ì´ {cls.total_students}ëª…ì˜ í•™ìƒì´ ì¬í•™ ì¤‘ì…ë‹ˆë‹¤."

    @classmethod
    def change_school_name(cls, new_name):
        cls.school_name = new_name
        return f"í•™êµëª…ì´ {cls.school_name}ë¡œ ë³€ê²½ë˜ì—ˆìŠµë‹ˆë‹¤."

    # ìŠ¤íƒœí‹± ë©”ì†Œë“œ (í•™êµì™€ ê´€ë ¨ëœ ìœ í‹¸ë¦¬í‹°)
    @staticmethod
    def is_passing_grade(score):
        return score >= 60

    @staticmethod
    def calculate_gpa(scores):
        if not scores:
            return 0
        return sum(scores) / len(scores)
````
